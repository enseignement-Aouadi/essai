Les fonctions sont représentées par : 
$$f(x)\ =\ 2.x$$
$$ g(x)\ =\ 5.x$$

```python
from numpy import arange, pi, sin, cos
from matplotlib.pyplot import *
import sys

debut = -5
fin = 5
pas = 0.001
x = arange(debut, fin, pas)
y = 2*x
z = -5*x
figure()
title("Fonction 2x ")
xlabel('x')
ylabel('y')
grid(True)
axis([x.min(),x.max(), y.min(), y.max()])

plot(x,y, 'g')
plot(x,z,'r')
show()
close()
```

La fonction inverse est représentée par : 
$$f(x)\ =\ \frac{2}{x}$$

```python
from numpy import *
import matplotlib.pyplot as plt
import sys
a = -5
b = 5 
x = []
y = []
pas = (b - a) / 200
abscisse = a
for k in range(0, 201):
    x.append(abscisse)
    y.append(1 / abscisse)
    abscisse += pas
axis([a, b, -10, 10])
grid(True)
plot(x, y)
show()
close()
```
